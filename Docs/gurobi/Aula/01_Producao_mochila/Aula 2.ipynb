```python
import gurobipy as gp

def le_dados(nome_arq):
    with open(nome_arq, "r") as f:
       linhas = f.readlines()
       qtd_itens = int(linhas[0].strip().split(" ")[0])
       capacidade = int(linhas[0].strip().split(" ")[1])
       valores = linhas[0].strip().split(" ")
        
       vet_pesos = [linha.strip().split(" ")[0] for linha in linhas][1:]
       vet_valores = [linha.strip().split(" ")[1] for linha in linhas][1:]
        
       #Rótulos dos items
       itens = list()
        
       for i in range(qtd_itens):
           rotulo = f"Item_{i}"
           itens.append(rotulo)
            
       #Cria o dicionario de pesos    
       pesos = dict()
       for idx, peso in enumerate(vet_pesos):
           rotulo = itens[idx]
           pesos[rotulo] = peso #O peso desse rótulo é peso
        
       #Cria o dicionario de valores    
       valores = dict()
       for idx, peso in enumerate(vet_valores):
           rotulo = itens[idx]
           valores[rotulo] = peso #O peso desse rótulo é peso
       return itens, capacidade, pesos, valores
```

```python
def solver_mochila(nome_arq):
    #Le os dados do arquivo
    itens, capacidade, pesos, valores = le_dados(nome_arq)
    m = gp.Model()
    m.setParam(gp.GRB.Param.OutputFlag,0) #Impede de imprimir os detalhes da solução
    #Inserir variáveis de decisão - Agora sao variaveis binarias - cria um dicionario de variaveis do gurobi, setando varias variaveis em um comando só
    x = m.addVars(itens, vtype = gp.GRB.BINARY)
    
    #Função Objetivo
    m.setObjective(
        gp.quicksum(x[i] * valores[i] for i in itens),
        sense = gp.GRB.MAXIMIZE
    )
    
    #Restrição de capacidade
    c1 = m.addConstr(
        gp.quicksum(x[i] * pesos[i] for i in itens) <= capacidade
    )
    
    m.optimize()
    print(m)
    
    return m.objVal
```

```python
arquivo = "Instancias/inst_000.txt"
resultado = solver_mochila(arquivo)
print(resultado)
```
