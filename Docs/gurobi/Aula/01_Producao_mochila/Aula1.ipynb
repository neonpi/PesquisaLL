```python
import gurobipy as gp
```

```python
#Criar o modeloo
m = gp.Model("Exemplo 1")

#Inserindo Variáveis de decisão
x1 = m.addVar()
x2 = m.addVar()
x3 = m.addVar()

#Define função objetivo
m.setObjective(12*x1+18*x2+22*x3, sense=gp.GRB.MAXIMIZE)

#Inserindo Restrições

c1 = m.addConstr(1.5 * x1 + 1.2 * x3 <= 120)
c2 = m.addConstr(2.2 * x2 + 1.4 * x3 <= 200)
c3 = m.addConstr(1.2 * x1 + 2 * x2 + 2.4 * x3 <= 250)

#Resolvendo o modelo.
m.optimize()
```

```python
#Acessando o valor das variaveis da solução
print("Produto A: ", x1.X)
print("Produto B: ", x2.X)
print("Produto C: ", x3.X)
```

```python
#Tempo de ocupação das máquinas (Precisa calcular o slack (Tempo de sobra) das máquina)
tempo_m1 = 120-c1.Slack
tempo_m2 = 200-c2.Slack
tempo_m3 = 250-c3.Slack
print("Máquina 1: ", tempo_m1)
print("Máquina 2: ", tempo_m2)
print("Máquina 3: ", tempo_m3)
```

```python
qtd_itens = 15
capacidade = 1000
# Gurobi trabalha com dicts

#Rótulos dos items
itens = list()

for i in range(qtd_itens):
    rotulo = f"Item_{i}"
    itens.append(rotulo)
    

vet_pesos = [65, 94, 119, 59, 149, 114, 57, 136, 100, 150, 122, 117, 120, 130, 133]
pesos = dict()
for idx, peso in enumerate(vet_pesos):
    rotulo = itens[idx]
    pesos[rotulo] = peso #O peso desse rótulo é peso

# print(pesos)

vet_valores = [455, 691, 833, 425, 1064, 758, 419, 914, 651, 966, 828, 827, 857, 837, 894]
valores = dict()
for idx, peso in enumerate(vet_valores):
    rotulo = itens[idx]
    valores[rotulo] = peso #O peso desse rótulo é peso
# print(valores)
```

```python
m_mochila = gp.Model()

#Inserir variáveis de decisão - Agora sao variaveis binarias - cria um dicionario de variaveis do gurobi, setando varias variaveis em um comando só
x = m.addVars(itens, vtype = gp.GRB.BINARY)

#Função Objetivo
m.setObjective(
    gp.quicksum(x[i] * valores[i] for i in itens),
    sense = gp.GRB.MAXIMIZE
)

#Restrição de capacidade
c1 = m.addConstr(
    gp.quicksum(x[i] * pesos[i] for i in itens) <= capacidade
)
m.optimize()
```

```python
for item in itens:
    # print(item, round(x[item].X))
    if round(x[item].X) == 1:
        print(item)

print("Valor total:", m.objVal)
print("Cabacidade Utilizada:",capacidade-c1.Slack)
```
