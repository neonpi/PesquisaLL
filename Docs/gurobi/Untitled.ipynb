```python
instance_file = open("../../instances/vrppl/25/C101_co_25.txt","r")

lines = [line.strip() for line in instance_file.readlines()]

#Salvando quantidade de clientes e quantidade e lockers
qty_customer = int(lines[0].split(" ")[0])
qty_locker = int(lines[0].split(" ")[1])

demmands = list()

## Salvando demmands de clients
for demmand in range(2,qty_customer+2):
    demmands.append(lines[demmand])


print(qty_customer, qty_locker)
print(demmands)
```
